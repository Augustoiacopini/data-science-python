# Preprocesamiento de  datos Titanic

exportamos los datos

```python
#librerias
import pandas as pd 
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
#conjunto de datos
prueba = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
```

```python
prueba.size
```

```python
train.size
```

```python
#concatenamos los datos
data = pd.concat([prueba, train], ignore_index=True)
```

```python
data.size
```

```python
data.columns
```

```python
data.head(10)
```

```python
data = data.drop(['Name'], axis=1)#elimina la columna nombres
```

## Mapeamos caracteristicas ordinales

```python
sex_maping = {'male':1, 'female':0}#Masculino = 1, femenino = 0
data['Sex'] = data['Sex'].map(sex_maping)#remplaza string por entero 
Embark_maping = {'C':1, 'Q':2, 'S':3}# Embarked C = Cherbourg = 1, Q = Queenstown = 2, S = Southampton = 3 puertos
data['Embarked'] = data['Embarked'].map(Embark_maping)
```

```python
data
```

```python
data.Pclass.unique()#Ticket class 1 = 1st, 2 = 2nd, 3 = 3rd

```

```python
data.dtypes
```

## Buscar valores nulos

```python
data.isnull().sum()
```

```python
data = data.drop(['Cabin'], axis=1)#elimina la columna cabina
```

```python
data = data.dropna(subset=['Survived'])#eliminamos las filas que contienen nulos dentro de Survived
```

```python
data = data.dropna(subset=['Embarked'])#eliminamos las filas que contienen nulos dentro de Embarked
```

```python
data.isnull().sum()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data = data.drop(['Ticket'], axis=1)#elimina la columna ticket
data = data.drop(['PassengerId'], axis=1)#elimina la columna PassengerId
```

```python
data.head(10)
```

### Imputamos valores ausentes en la columna Age

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(data.values)
imp_t = imp.transform(data.values)
```

```python
imputad = pd.DataFrame(imp_t)
imputad.isnull().sum()
```

```python
imputad[2].values
```

```python
data['Age'].values
```

```python
data['Age'] = imputad[2].values
```

```python
data.isnull().sum()
```
